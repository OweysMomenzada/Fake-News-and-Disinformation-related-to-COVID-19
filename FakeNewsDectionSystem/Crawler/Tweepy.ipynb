```python
import pandas as pd
import tweepy

'''
Function to display the data of each extracted tweet
'''
def printtweetdata_simple(ith_tweet):
  print()
  print(f"Username:{ith_tweet[0]}")
  print(f"Tweet Text:{ith_tweet[1]}")
  print("\n------------------------------------")

 
'''
Get tweets by passed parameters in simple format
'''
def get_tweets(words, date_since, numtweet):
	
	# Pandas dataframe
	db = pd.DataFrame(columns=['username','text'])
	
	tweets = tweepy.Cursor(api.search, q=words, lang="en",
						since=date_since, tweet_mode='extended').items(numtweet)
	
	list_tweets = [tweet for tweet in tweets]
	
	for tweet in list_tweets:
		username = tweet.user.screen_name
		# Get text
		try:
			text = tweet.retweeted_status.full_text
		except AttributeError:
			text = tweet.full_text
		
		ith_tweet = [username, text]
		db.loc[len(db)] = ith_tweet
		
		printtweetdata_simple(ith_tweet)
  
  # Save our dataframe as a CSV file.
	filename = 'scraped_tweets.csv'
	db.to_csv(filename)

	
# Credentials
consumer_key = "<>"
consumer_secret = "<>"
access_token = "<>"
access_token_secret = "<>"

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
	
# Enter search hashtag, date and num of tweets
print("Enter search word to search for:")	
words = input()
print("Enter DATE since tweets are posted: [yyyy-mm--dd]")
date_since = input()
print("Enter AMOUNT of tweets to extract:")	
numtweet = int(input())

# Get tweets
get_tweets(words, date_since, numtweet)
```
